```python
#Packages
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from smt.sampling_methods import LHS
import pandas as pd
```

```python
#Sampling --- Temperature / Duration
#dont run - lhs is quasi-random and it will get different results
limits = np.array([[870, 1125], [1200, 9000]])
sampling = LHS(xlimits=limits, random_state=42)

test_size = 10

Xtest = sampling(test_size)

print(Xtest)
```

```python
limits = np.array([[870, 1125], [1200, 9000]])
sampling = LHS(xlimits=limits, random_state=42)

test_sampling = 20

X_test = sampling(test_sampling)

print(X_test)
```

```python
df_train = pd.read_csv(r'C:\Users\mikep\Desktop\project\data\final\Xtrain.csv')
df_train
```

```python
#Figure
plt.rcParams.update({'font.size': 15})
plt.rcParams["figure.figsize"] = (12, 10)

ax = df_train.plot.scatter(x='Temperature', y='Duration', color='white', s=100, linewidth=1, marker='o', edgecolor='black')
df_test.plot.scatter(ax=ax, x='Temperature', y='Duration', color='black', s=100, linewidth=1, marker='o', edgecolor='black')

plt.title('Latin Hypercube Sampling', fontsize=30, pad=20)

plt.xlabel("Temperature (Kelvin)", fontsize=25, labelpad=20)
plt.ylabel("Annealing Time (seconds)", fontsize=25, labelpad=20)

label_size = 20
plt.rcParams['xtick.labelsize'] = label_size
plt.rcParams['ytick.labelsize'] = label_size
plt.tick_params(axis="both",size=15, direction="in", pad=8)

plt.legend('')

plt.grid(which='major', axis='both', linestyle='--', linewidth=1.5)
#plt.savefig(r'C:\Users\mikep\Desktop\thesis\figures\sampling\sampling_train.png', dpi=600)
plt.show()
```

```python
#Packages
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from smt.sampling_methods import LHS
import pandas as pd
```

```python
#Sampling --- Temperature / Duration
limits = np.array([[870, 1125], [1200, 9000]])
sampling = LHS(xlimits=limits)

num = 20
x = sampling(num)

Xtrain = x[0:10]
Xtest = x[10:20]

print(x)
print(x.shape, Xtrain.shape, Xtest.shape)
```

```python
#Figure
plt.rcParams["figure.figsize"] = (15, 12)

plt.plot(Xtrain[:, 0], Xtrain[:, 1], "o", c='white', markersize=10, markeredgecolor='black')
plt.plot(Xtest[:, 0], Xtest[:, 1], "o", c='black', markersize=10, markeredgecolor='black')

plt.title('Latin Hypercube Sampling', fontsize=30, pad=20)

plt.xlabel("Temperature (Kelvin)", fontsize=25, labelpad=20)
plt.ylabel("Annealing Time (seconds)", fontsize=25, labelpad=20)

label_size = 20
plt.rcParams['xtick.labelsize'] = label_size
plt.rcParams['ytick.labelsize'] = label_size
plt.tick_params(axis="both",size=15, direction="in", pad=8)

plt.grid(which='major', axis='both', linestyle='--', linewidth=1)
plt.savefig(r'C:\Users\mikep\Desktop\thesis\figures\sampling.png', dpi=200)
plt.show()
```

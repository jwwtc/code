```python
#Packages
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from smt.sampling_methods import LHS
import pandas as pd
```

```python
#Sampling --- Temperature / Duration
limits = np.array([[870, 1125], [1200, 9000]])
sampling = LHS(xlimits=limits)

test_size = 10

Xtest = sampling(test_size)

print(Xtest)
```

```python
df_train = pd.read_csv(r'C:\Users\mikep\Desktop\project\data\samples_train.csv')
df_train
```

```python
#Figure
plt.rcParams.update({'font.size': 15})
plt.rcParams["figure.figsize"] = (12, 10)

df_train.plot(x='Temperature', y='Duration', color='white', linewidth=3, marker='o', markersize=10, markeredgecolor='black')
#plt.plot(Xtest[:, 0], Xtest[:, 1], "o", c='black', markersize=10, markeredgecolor='black')

plt.title('Latin Hypercube Sampling', fontsize=30, pad=20)

plt.xlabel("Temperature (Kelvin)", fontsize=25, labelpad=20)
plt.ylabel("Annealing Time (seconds)", fontsize=25, labelpad=20)

label_size = 20
plt.rcParams['xtick.labelsize'] = label_size
plt.rcParams['ytick.labelsize'] = label_size
plt.tick_params(axis="both",size=15, direction="in", pad=8)

plt.legend('')

plt.grid(which='major', axis='both', linestyle='--', linewidth=1.5)
plt.savefig(r'C:\Users\mikep\Desktop\thesis\figures\sampling\sampling_train.png', dpi=600)
plt.show()
```

```python
#Figure
plt.rcParams.update({'font.size': 15})
plt.rcParams["figure.figsize"] = (12, 10)

df_train.plot(x='Temperature', y='Duration', color='white', linewidth=3, marker='o', markersize=10, markeredgecolor='black')
plt.plot(Xtest[:, 0], Xtest[:, 1], "o", c='black', markersize=10, markeredgecolor='black')

#plt.title('Latin Hypercube Sampling', fontsize=30, pad=20)

plt.xlabel("Temperature (Kelvin)", fontsize=25, labelpad=20)
plt.ylabel("Annealing Time (seconds)", fontsize=25, labelpad=20)

label_size = 20
plt.rcParams['xtick.labelsize'] = label_size
plt.rcParams['ytick.labelsize'] = label_size
plt.tick_params(axis="both",size=15, direction="in", pad=8)

plt.legend('')

plt.grid(which='major', axis='both', linestyle='--', linewidth=1.5)
plt.savefig(r'C:\Users\mikep\Desktop\thesis\figures\sampling\sampling_test.png', dpi=600)
plt.show()
```

```python
#Packages
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from smt.sampling_methods import LHS
import pandas as pd
```

```python
#Sampling
limits = np.array([[870, 1125], [1200, 9000]])
sampling = LHS(xlimits=limits)

train_size = 20
test_size = 10

Xtrain = sampling(train_size)
Xtest = sampling(test_size)
```

```python
df=pd.DataFrame(Xtrain)
df.columns = ['Temperature', 'Duration']
df_test=pd.DataFrame(Xtest)
df_test.columns = ['Temperature_t', 'Duration_t']
```

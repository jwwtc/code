```python
#Packages
import warnings
warnings.filterwarnings('ignore')
import glob
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from PIL import Image
from pymks import (
    generate_multiphase,
    plot_microstructures,
    PrimitiveTransformer,
    TwoPointCorrelation,
    FlattenTransformer
)
```

```python
grain_sizes = [(5, 15), (10, 40), (15, 15), (5, 30)]
seeds = [10, 99, 4, 36]

data_synth = np.concatenate([
    generate_multiphase(shape=(10, 100, 100), grain_size=grain_size,
        volume_fraction=(0.5, 0.5),
        percent_variance=0,
        seed=seed
    )
    for grain_size, seed in zip(grain_sizes, seeds)
])
data_synth.shape
```

```python
plot_microstructures(*data_synth[::5], cmap='gray', colorbar=False)
```

```python
data_disc = PrimitiveTransformer(n_state=2, min_=0.0, max_=1.0).transform(data_synth)
data_disc.shape
```

```python
data_corr = TwoPointCorrelation(
    periodic_boundary=True,
    cutoff=10,
    correlations=[(0, 0)]
).transform(data_disc)
```

```python
data_corr.shape
```

```python
plot_microstructures(
    data_corr[0, :, :, 0],
    data_corr[2, :, :, 0],
    titles=['1st Microstructure - Correlation [0, 0]', '3rd Microstructure - Correlation [0, 0]'],
    showticks=True
)
```
